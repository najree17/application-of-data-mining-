```python
# date parser
d_parser = lambda date: pd.to_datetime(date, format=' %d-%m-%Y')

# load the dataset
afit_data = pd.read_csv(
    'AFIT.csv', parse_dates=[1, 2],
    index_col=[0], date_parser=d_parser, engine='python'
)
```

```python
afit_data
```

```python
consumption = []

for i in range(len(afit_data.values)):
    x = abs((afit_data.values[i][0] - afit_data.values[i][1]).days) * 24
    y = np.random.multinomial(n=afit_data.values[i][2], pvals=[1/x] * x)
    consumption += list(y)
```

```python
len(consumption)
```

```python
consumption[:10]
```

```python
afit_dataframe['ENERGY CONSUMPTION'] = [i for i in consumption]
```
